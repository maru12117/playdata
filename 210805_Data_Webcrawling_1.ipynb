## 웹 크롤링(Webcraling) 기초

bs4의 BeautifulSoup을 많이 이용하기도 하지만,

웹크롤러(ChromeDriver)를 통해 자동화, 매크로(Macro) 작업을 하고 싶기 때문에 Selenium 프레임워크를 사용함

```python
from selenium import webdriver

driver=webdriver.Chrome("C:/Users/pc/chromedriver/chromedriver_win32/chromedriver.exe")

url ='https://www.naver.com/'
driver.get(url) #네이버 홈페이지에 접속

html=driver.page_source #웹페이지의 html정보를 가져옴
```

```python
html
```

```python
from bs4 import BeautifulSoup

soup=BeautifulSoup(html, 'html.parser')
soup
```

```python
tags_span=soup.select('span') #모든 span은 다나옴
tags_p = soup.select('p')
```

```python
p_desc=soup.select('p.desc') #<p>에서의 조건을 추가함  class가 desc인 것을 조건으로 추가
p_desc
```

```python
class_desc=soup.select('.desc') #클래스가 desc인것들
class_desc
```

```python
from selenium import webdriver

driver1=webdriver.Chrome("C:/Users/pc/chromedriver/chromedriver_win32/chromedriver.exe")

driver1.get("https://movie.daum.net/main")

html=driver1.page_source
```

```python
from bs4 import BeautifulSoup

soup=BeautifulSoup(html, "html.parser")
soup
```

```python
class_link=soup.select(".link_relservice")
class_link
```

```python
class_link[0] #인덱스 사용도 가능함
```

```python
for tag in class_link:
    print(tag)
```

```python
for tag in class_link:
    print(tag.text) #.text를 활용하여 선택한 정보만 가져오기
```

```python
for tag in class_link:
    print(tag["href"]) #속성값 추출하기
```
