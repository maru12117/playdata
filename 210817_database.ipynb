## 데이터베이스 vs. DBMS
  - 데이터의 집합, 데이터의 저장소 vs. 소프트웨어
  - 데이터 파일 vs. 프로그램 파일


## 데이터베이스 특징
  1) 데이터의 무결성(Integrity)
    ⇐ 제약조건(Constraint)으로 데이터 무결성을 지원
    
  2) 데이터의 독립성(Isolation)
    ⇐ DBMS에서 구현
    
  3) 보안(Security)
    ⇐ 권한(Privileges)
    
  4) 데이터 중복 최소화
    ⇐ 관계형데이터베이스 ⇐ 관계 대수
    
  5) 응용 프로그램 제작 및 수정이 쉬워짐
  
  6) 데이터의 안전성 향상
    ⇐ 백업(Backup), 복원(Restore), 복구(Recover = Restore + Redo Log)


## 관계형데이터베이스(Relational DB)
  - 관계대수(Relational Algebra)에 기반하여 설계된 DB
  - Relation(관계) vs. Relationship(관계)
   ⇒ 데이터의 집합 vs. 관련 있는 릴레이션을 표현
   ⇒ 테이블(표) vs. 참조제약조건(Foreign Key)

  cf.) 테이블: 데이터의 집합   (집합적 사고)
      ex) 사원 테이블: employees
          부서 테이블: departments
          고객 테이블: customers
          도시 테이블: cities



데이터베이스의 발전 순서

오프라인 ⇒ 파일시스템 ⇒ 데이터베이스 관리시스템(DBMS)

## SQL(Structured Query Language)개요

  - 기능이 이미 갖춰져 있어, 데이터 삽입 및 삭제, 검색 등이 용이함
  - 데이터베이스 및 데이터베이스 객체 생성, 변경, 삭제가 가능
  - 다만, 수많은 기능에 대한 sql명령과 수많은 옵션을 익혀야 함
  
## SQL과 다른 언어의 차이점

  - 대소문자를 구별하지않음
  - 변수가 없음(예외로 바인드 변수)
  - 반복문이 없음
  - 조건문이 없음
  - 예외처리가 없음
  
## Daemon(Unix-like), Service(Windows)

## 시스템 설계와 모델링

정보시스템을 구축하기 위해서는 일반적으로 분석, 설계, 구현, 시험, 유지보수 등 5가지의 단계를 거침
 - 분석 : 무엇을  구축할지..
 - 설계 : 어떻게..구축할지
 - 구현
 - 시험
 - 유지보수
 
데이터베이스 모델링과 필수용어
 - 분석과 설계 단계에서 중요하고 필수인 단계
 - 커뮤니케이션의 매개체
 - 현실 세계의 데이터를 DBMS에 어떻게 옮겨놓을 것인지 결정하는 과정
 
 키(key)
  - 로우를 다른 로우와 구별(식별)할 수 있는 컬럼(의 값)
  - 자연키(Natural Key) vs. 합성키(Synthetic Key), 인공키(Artificial Key)
  
 기본 키(Primary Key)
  - 키 중에서 선택
  - 특징
  
    1) 유일(Unique) : 주민등록번호, 학번
    
    2) 값이 반드시 존재(Not Null)
    
    3) 테이블 당 하나만 존재함, 대신 필요하다면 2개 이상의 컬럼을 조합하여 하나의 PK로 설정이 가능함
    
    
   - 키 중에서 선택(잘 변하지 않는 후보 키, 개인정보 또는 보안상 중요하지 않은 후보 키)

## Query

 1) 광의의 의미 : 모든 SQL문
 2) 협의의 의미 : SELECT 문
  cf) query and dml => SELECT and INSERT, UPDATE, DELETE
  
## SQL 문의 구분
 1) DML(Data Manipulation Language)
 
   => 대상:DATA
   
   => SQL: Insert, Updata, Delete, Merge...Select
   
 2) DDL(Data Definition Laguage)
 
   => 대상 : Database Objects
   
   => SQL: Create, Alter, Drop...
   
 3) DCL(Data Control Language)
 
   => 권한, 트랜잭션,..
   
   => SQL : Grant 권한 to....
           Revoke 권한 From...., Commit, Rollback, Savepoint
