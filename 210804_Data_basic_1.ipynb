## 2개의 테이블을 합치기(merge, append)

```python
sample1
```

```python
code_master=pd.read_excel("C:/Users/pc/Main/Data_python/2_Data_Analysis_Basic/files/sample_codemaster.xlsx")
code_master
```

```python
sample1_code = pd.merge(left=sample1,
                       right=code_master,
                       how="left",
                       left_on="국적코드",
                       right_on="국적코드")
sample1_code #how="left" 왼쪽 칼럼을 기준으로
```

```python
sample1_code_inner = pd.merge(left=sample1,
                       right=code_master,
                       how="inner",
                       left_on="국적코드",
                       right_on="국적코드")

sample1_code_inner #두 테이블의 칼럼이 공통된 것만 추출하겠다..how="inner"
```

```python
# 데이터를 월별 파일로 관리하는 경우가 많기 떄문에....여러개의 excel파일을 병합함(append)

sample2=pd.read_excel("C:/Users/pc/Main/Data_python/2_Data_Analysis_Basic/files/sample_2.xlsx", header=1, skipfooter=2, usecols="A:C")
sample2
```

```python
sample2["기준년월"]="2021-9"
sample2
```

```python
sample2_code=pd.merge(left = sample2,
                     right = code_master,
                     how="left",
                     left_on="국적코드",
                     right_on="국적코드")
sample2_code
```

```python
sample1_code
```

## sample1_code와 sample2_code를 append로 병합함(행 병합)

```python
sample = sample1_code.append(sample2_code, ignore_index=True) #인덱스 무시하고 정렬함
sample
```
