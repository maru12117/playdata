## 데이터 전처리 과정을 함수로 만들어서 모든 데이터를 한번에 함

```python
def create_kto_data(yy,mm):
    file_path="C:/Users/pc/Main/Data_python/4_Tourists_Event/files/kto_{}{}.xlsx".format(yy,mm)
    
    df = pd.read_excel(file_path, header=1, skipfooter=4, usecols="A:G")
    
    #기준년월 칼럼 추가
    df["기준년월"]='{}-{}'.format(yy,mm)
    
    #국적칼럼에서 대륙을 제거함
    continents_list=["아시아주","미주","구주","대양주","아프리카주","기타대륙","교포소계"]
    continents_list
    condition=(df["국적"].isin(continents_list)==False)
    #인덱스 다시 설정
    df_country = df[condition].reset_index(drop=True)
    
    #대륙 칼럼 추가
    continents = ["아시아"]*25+["아메리카"]*5+["유럽"]*23+["오세아니아"]*3+["아프리카"]*2+["기타대륙"]*1+["교포"]
    df_country["대륙"] = continents
    
    #국가별 관광객비율(%) 칼럼 추가
    df_country["관광객비율(%)"]=round(df_country["관광"]/df_country["계"]*100,1)
    
    #"전체비율(%)"칼럼 추거
    tourist_sum=sum(df_country["관광"])
    df_country["전체비율(%)"] = round(df_country["관광"]/tourist_sum*100,1)
    
    #결과 출력
    return (df_country)
    
```

```python
kto_test=create_kto_data(2018,12)
kto_test.head()
```

## 히트맵 그래프 그리기

```python
df.head()
```

```python
# 기준년월도를 슬라이싱 하여....."년도"와 "월"의 column을 만들기..str.slice()함수 활용

df["연도"]=df["기준년월"].str.slice(0,4)
df["월"]=df["기준년월"].str.slice(5,7)
df.head()
```

```python
condition=(df["국적"]=="중국")
df_filter=df[condition]
df_filter.head()
```

```python
#pivot_table를 사용하여 

df_pivot = df_filter.pivot_table(values="관광", index="연도",columns="월")
df_pivot
```

```python
#히트맵 그래프는 seaborn을 import 해야하나..matplotlib에 종속되기떄문에...함꼐 import 해야한다!

import matplotlib.pyplot as plt
import seaborn as sns

#그래프 크기 설정
plt.figure(figsize=(16,10))

#히트맵 그래프 그리기
sns.heatmap(df_pivot, annot=True, fmt='.0f', cmap="rocket_r")

#그래프 타이틀
plt.title("중국 관광객 히트맵")
plt.show()
```

```python
for cnt in cntry_list:
    condition=(df["국적"]==cnt)
    df_filter=df[condition]
    
    df_pivot=df_filter.pivot_table(values="관광", index="연도",columns="월")
    
    plt.figure(figsize=(16,10))
    sns.heatmap(df_pivot, fmt=".0f", cmap="rocket_r",annot=True)
    
    plt.title("{} 관광객 히트맵".format(cnt))
    
    plt.show()
```
