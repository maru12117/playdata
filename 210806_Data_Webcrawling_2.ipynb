```python
from selenium import webdriver

driver=webdriver.Chrome("C:/Users/pc/chromedriver/chromedriver_win32/chromedriver.exe")

url = "https://www.melon.com/chart/"
driver.get(url)
```

```python
from bs4 import BeautifulSoup

html=driver.page_source
soup=BeautifulSoup(html,"html.parser")
soup
```

```python
songs=soup.select("table>tbody>tr")
songs[0]
```

```python
title=songs[0].select('div.ellipsis.rank01>span>a')[0].text
title
```

```python
for i in range(10):
    print(songs[i].select('div.ellipsis.rank01>span>a')[0].text)
```

## 가수 이름 찾기

```python
song=soup.select("table>tbody>tr")
song[0]
```

```python
singer=song[0].select('div.ellipsis.rank02 > a')
singer[0].text
```

```python
for i in range(10):
    title=songs[i].select('div.ellipsis.rank01>span>a')[0].text
    singer=song[i].select('div.ellipsis.rank02 > a')[0].text
    print(f"{i+1}위 {title}, 가수 {singer}")
```

## 결과를 엑셀 파일에 저장하기

```python
song_list=[] #리스트를 생성하여 넣음
for i in range(10):
    title=songs[i].select('div.ellipsis.rank01>span>a')[0].text
    singer=song[i].select('div.ellipsis.rank02 > a')[0].text
    song_list.append(['Melon',i+1,title, singer])
song_list 
```

```python
import pandas as pd

columns =["서비스","순위","노래 제목","가수"]
pd_data = pd.DataFrame(song_list, columns=columns)
pd_data
```

```python
pd_data.to_excel("C:/Users/pc/Main/Data_python/melon.xlsx", index=False) #인덱스 없이...저장하기
```

## 지니 순위 크롤링해서 엑셀파일로 변형하기

```python
driver=webdriver.Chrome("C:/Users/pc/chromedriver/chromedriver_win32/chromedriver.exe")
driver.get("https://www.genie.co.kr/chart/top200?ditc=D&rtm=N")
```

```python
html=driver.page_source
soup=BeautifulSoup(html,"html.parser")
```

```python
songs=soup.select("table>tbody>tr")
```

```python
title=songs[0].select('a.title.ellipsis')[0].text.strip()
title
```

```python
singer=songs[0].select('a.artist.ellipsis')[0].text
singer
```

```python
song_list=[]
for i in range(10):
    title=songs[i].select('a.title.ellipsis')[0].text.strip()
    singer=songs[i].select('a.artist.ellipsis')[0].text
    song_list.append(['지니',i+1,title,singer])
song_list
```

```python
data1=pd.DataFrame(song_list, columns=columns)
data1.to_excel("C:/Users/pc/Main/Data_python/genie.xlsx", index=False)
```
