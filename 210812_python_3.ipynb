```python
##시간을 포맷에 맞춰서 출력하기

now.strftime("%Y.%m.%d %H:%M:%S")
```

```python
"{}년 {}월 {}일 {}시 {}분 {}초".format(now.year, now.month, now.day, now.hour, now.minute, now.second)
```

```python
now.strftime("%Y{}.%m{}.%d{} %H{}:%M{}:%S{}").format(*"년월일시분초")
```

```python
#특정 시간 이후의 시간 구하기(시간 더하기)

after=now + datetime.timedelta(\
                        weeks=1,\
                        days=1,\
                        hours=1,\
                              minutes=1,\
                              seconds=1)

after
```

```python
after.strftime("%Y{}.%m{}.%d{} %H{}:%M{}:%S{}").format(*"년월일시분초")
```

```python
import os


#파일과 폴더를 모두 출력함
output=os.listdir(".")
output
```

```python
os.mkdir("hello")
```

```python
#폴더와 파일을 구분하기

for path in output:
    if os.path.isdir(path): #파일인지 확인함.
        print("폴더: ", path)
    else:
        print("파일: ", path)
```

```python
os.listdir(".")
```

```python
#최근 결과
_[2]
```

```python
# 재귀 구성으로 현재 폴더 내부에 있는 모든 파일을 탐색하도록 코드를 작성

def read_folder(path):
    output=os.listdir(path)
    
    for item in output:
        if os.path.isdir(item):
            read_folder(path+"/"+item)
        else:
            print("파일:",item)
    
    
read_folder(".")
```
