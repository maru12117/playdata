## BeautifulSoup 모듈 활용하기

```python
from bs4 import BeautifulSoup
from urllib import request

target=request.urlopen("http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108")
soup=BeautifulSoup(target,"html.parser")
soup
```

```python
for location in soup.select("location"):
    print("도시 : ",location.select_one("city").text)
    print("날씨 : ", location.select_one("wf").text)
    print("최저기온 : ", location.select_one("tmn").text)
    print("최고기온 : ", location.select_one("tmx").text)
    print()
```

## flask

```python
from flask import Flask
from bs4 import BeautifulSoup
from urllib import request

app = Flask(__name__)

@app.route("/")
def hello():
    
    target=request.urlopen("http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108")
    soup=BeautifulSoup(target,"html.parser")
    
    output=""
    for location in soup.select("location"):
        output += "<h3>{}</h3>".format(location.select_one("city").text)
        output += "날씨 :{}</br>".format(location.select_one("wf").text)
        output += "최저/최고기온 : {}/{}".format(\
                                     location.select_one("tmn").text,\
                                     location.select_one("tmx").text)
        output+="<hr/>"

    return output

if __name__ == "__main__":
    app.run()
```

## 함수 데코레이터

```python
@함수 데코레이터를 생성함
def test(function):
    def wrapper():
        print("인사가 시작되었습니다")
        function() #@밑에 있는 것들을..
        print("인사가 종료되었습니다")
    return wrapper

#데코레이터를 붙여 함수를 만든다...
@test
def hello():
    print("hello")

#함수 호출
hello()
```
